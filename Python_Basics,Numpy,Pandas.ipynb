A Fibonacci sequence is a series of numbers where:

Each number is the sum of the two numbers before it.
The Tribonacci sequence is like the Fibonacci sequence, but instead of adding the last two numbers, you add the last three numbers.

```python
# Generate first 10 Fibonacci numbers
fib = [0, 1]  # starting values

for i in range(8):  # we already have 2 numbers, so do this 8 more times
    next_number = fib[-1] + fib[-2]  # sum of last two numbers
    fib.append(next_number)

print(fib)
```

```python
todo_list = []

def add_task():
    task = input("Enter your task: ")
    todo_list.append(task)
    print("Task added.")

def view_tasks():
    if not todo_list:
        print("No tasks added.")
    else:
        print("\nYour To-Do List:")
        for i, task in enumerate(todo_list, 1):
            print(f"{i}. {task}")

def delete_task():
    view_tasks()
    if todo_list:
        task_num = int(input("Enter task number to delete: "))
        if 0 < task_num <= len(todo_list):
            removed = todo_list.pop(task_num - 1)
            print(f"Deleted task: {removed}")
        else:
            print("Invalid task number.")

def main():
    while True:
        print("\nTo-Do List Options:")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Delete Task")
        print("4. Exit")

        choice = input("Choose an option: ")

        if choice == '1':
            add_task()
        elif choice == '2':
            view_tasks()
        elif choice == '3':
            delete_task()
        elif choice == '4':
            print("Goodbye!")
            break
        else:
            print("Invalid choice.")

main()
```

```python
print("Welcome to neocloud Atm withdrawal")

# We created a varible here that stores all the data of the users
# The varible is what we call a nested dictionary in python

data = {"tony":{"password":1234 , "balance":1000}}


def logIn():
  '''
  This function authenticates and logins in user
  '''
  user_name = input("Input your name: ")
  if user_name in data:
    print("Account exist ")
    user_password = input("Enter your Password: ")

    if int(data[user_name]["password"]) == int(user_password):
      print("Verified")
      return user_name

    else:
      retryPassword(user_name)
      return user_name


  else:
    new_user_age = input("Enter a Password To Create Account: ")
    data[user_name] = {}
    data[user_name]["password"] = int(new_user_age)
    data[user_name]["balance"] = 0
    print("Account created")
    logIn()

    return user_name





def retryPassword(user_name):
  '''
  This function retries the password if incorrect

  '''
  attempts = 0
  while(attempts < 3):
    attempts += 1
    if attempts == 3:
      print("Too many attempts")
    else:
      print("Incorrect password, Try again")
      user_password = input("Enter your Password: ")
      if int(data[user_name]["password"]) == int(user_password):
        print("Verified")
        break





def deposit():
  '''
  This function deposits money into the account

  '''
  user_name = logIn()
  amount = input("Input amount to deposit: ")

  data[user_name]["balance"] =  int(data[user_name]["balance"]) +  int(amount)
  print(f'Deposit successful , your new balance is {data[user_name]["balance"]}')



def withdraw():
  '''
  This function withdraws money from the account

  '''
  user_name = logIn()
  amount = input("Input amount to withdraw: ")
  if int(data[user_name]["balance"]) < int(amount):
    print("Insufficient balance")
  else:
    data[user_name]["balance"] =  int(data[user_name]["balance"]) -  int(amount)
    print(f'Withdrawal successful , your new balance is {data[user_name]["balance"]}')


def viewBalance():
  '''
  This function views the balance of the account

  '''
  user_name = logIn()
  print(f'Your balance is {data[user_name]["balance"]}')

# This print just output what actions a user can perform in the atm

print(
    "  \n1: To create Account \n2: To deposit \n3: To Withdraw \n4: To view balance \n5: To end "
  )
endProcess = False
while(endProcess == False):
  # we request the user to select an action with this line of code
  action = input( "\nSelect an action with the action Number: " )

  # we use the conditonal statement to check if the action is valid and call the right function based on the action selected

  if (int(action) == 1):
    # we create a new account or log in if 1 is selected
    print("create Account or logIn")
    logIn()

  elif (int(action) == 2):
    # we deposit money into out account by calling the deposit function  if action  2  is selected
    deposit()

  elif (int(action) == 3):
    # we withdraw money from account here by calling the withdraw function if 3 is selected
    withdraw()

  elif (int(action) == 4):
    # This is just to view out current balance by selection 4 as action
    viewBalance()

  else:
    if (int(action) == 5):
      #This ends the process by selecting 5
      endProcess = True
      print("Thank you for using neocloud Atm")
    else:
      print("Invalid action")
      break
```

```python
# Smart Profile Builder for Internship Applications
#Pseudocode
#start
# ask user for input ie fullname, age, email and interest
# ensure the user enters numbers only
# remove white spaces from the input, make email lowercase, capitalize the first letter of each name, split the interests into a list and remove whitespaces
# remove duplicates and sort the list alphabetically
# determine age group using the parameters given
# Display name in reversed and uppercased using string methods
# display age groups
# show number of interests
# Show the first and last interests
# Sort the interests and display
# add eligibility for the internship
# ask if the user wants to add another input, if yes, repeat the process
# If not, acknowledge the user and end the program
# stop



while True:
    full_name = input("Enter your full name: ")
    age_input = input("Enter your age: ")
    email = input("Enter your email address: ")
    interests_input = input("Enter 3 interests (comma-separated): ")

    if not age_input.isdigit():
        print("Invalid age. Please enter numbers only.")
        continue
    age = int(age_input)

    full_name = full_name.strip().title()
    email = email.strip().lower()

    interests = interests_input.split(",")
    cleaned_interests = []

    for item in interests:
        item = item.strip().title()
        if item not in cleaned_interests:
            cleaned_interests.append(item)

    cleaned_interests.sort()

    if age >= 13 and age <= 17:
        age_group = "Teen"
    elif age >= 18 and age <= 25:
        age_group = "Young Adult"
    else:
        age_group = "Other"

    print("Profile Summary")
    reversed_name = full_name[::-1].upper()
    print(f"Name (reversed & uppercase): {reversed_name}")

    print(f"Age Group: {age_group}")
    print("Number of Interests: {}".format(len(cleaned_interests)))

    if len(cleaned_interests) > 0:
        print("First Interest: {}".format(cleaned_interests[0]))
        print("Last Interest: {}".format(cleaned_interests[-1]))

    print("Sorted Interests: {}".format(", ".join(cleaned_interests)))

    if age >= 18:
        print("You are eligible for internship.")
    else:
        print("You are not eligible for internship yet.")

    again = input("\nDo you want to enter another profile? (yes/no): ").strip().lower()
    if again != "yes":
        print("Thank you for using the Smart Profile Builder.")
        break

```

```python
# A student departure system
students = {}
DEAN_PASSWORD = "neocloud231"

def student_login():
  name = input("enter your full name: ").strip().title()
  student_id = input("enter your student id: ").strip()

  if student_id not in students:
    print("no account found creating new student record...")
    students[student_id] = {
      "name": name,
      "exeat_status" : "on campus",
      "history":[]
    }
  else:
    print(f"welcome back {students[student_id]['name']}")

  return student_id

def request_exeat(student_id):
  if student_id not in students:
    print("Please login first.")
    return

  if students[student_id]["exeat_status"] == "off campus":
    print("you already have an active exeat. Return first.")
    return

  reason = input("why are you requesting an exeat?").strip()
  students[student_id]["exeat_status"] = "pending approval"
  students[student_id]["history"].append({"request":reason,"status":"pending"})

  print("exeat request submitted.awaiting dean approval.")

def approve_exeat():
  password = input("enter dean password: ").strip()
  if password != DEAN_PASSWORD:
    print("incorrect password.")
    return

  approved_count = 0
  for student_id,info in students.items():
    if info["exeat_status"] == "pending approval":
      print(f"\n{info['name']} ({student_id}) requested: {info['history'][-1]['request']}")
      decision = input("Approve? (yes/no): ").strip().lower()
      if decision == "yes":
        info["exeat_status"] = "off campus"
        info["history"][-1]["status"] = "approved"
        print("approved")
        approved_count += 1
      else:
        info["exeat_status"] = "on campus"
        info["history"][-1]["status"] = "denied"
        print("denied")
  if approved_count == 0:
      print("No pending exeat requests.")

def return_to_campus(student_id):
  if student_id not in students:
    print("Please login first.")
    return

  if students[student_id]["exeat_status"] != "off campus":
    print("you are not currently out of campus")
    return

  students[student_id]["exeat_status"] = "on campus"
  print("welcome back.you have successfully returned")

def view_history(student_id):
  if student_id not in students:
    print("Please login first.")
    return

  print("\nyour exeat history:")
  if not students[student_id]["history"]:
    print("No exeat history found.")
  else:
    for record in students[student_id]["history"]:
      print(f"{record['request']}: {record['status']}")

def security_view():
  print("\nstudents currently off campus:")
  off_campus_students = 0
  for student_id,info in students.items():
    if info["exeat_status"] == "off campus":
      print(f"{info['name']} ({student_id})")
      off_campus_students += 1
  if off_campus_students == 0:
      print("No students are currently off campus.")

def search_student():
  student_id = input("enter your student id to search: ").strip()
  if student_id in students:
    name = students[student_id]["name"]
    status = students[student_id]["exeat_status"]
    print(f"{name} is currently: {status}")
  else:
    print("student not found")

def drive():
  current_user = None  # Initialize current_user to None
  while True:
    print("\n====student departure system(sds)====")
    print("1. student login")
    print("2. request exeat")
    print("3. return to campus")
    print("4. view exeat history")
    print("5. dean approval panel")
    print("6. security view")
    print("7. search student")
    print("8. exit")

    choice = input("select an option(1-8):").strip()

    if choice == "1":
      current_user = student_login()
    elif choice == "2":
      if current_user:
        request_exeat(current_user)
      else:
        print("Please login first.")
    elif choice == "3":
      if current_user:
        return_to_campus(current_user)
      else:
        print("Please login first.")
    elif choice == "4":
      if current_user:
        view_history(current_user)
      else:
        print("Please login first.")
    elif choice == "5":
      approve_exeat()
    elif choice == "6":
      security_view()
    elif choice == "7":
      search_student()
    elif choice == "8":
      print("thank you for using SDS.Goodbye.")
      break
    else:
      print("invalid choice.")

drive()
```

Data Science terms:

1. Data Science â€“ The field that uses statistics, programming, and domain knowledge to extract insights from data.

2. Dataset â€“ A structured collection of data, often in rows (records) and columns (features).

3. Feature â€“ A variable or column in a dataset that represents an attribute of the data.

4. Target Variable â€“ The outcome or label you want to predict in supervised learning.

5. Supervised Learning â€“ Machine learning where models are trained with labeled data.

6. Unsupervised Learning â€“ Machine learning without labels; the model finds patterns or groups in the data.

7. Model â€“ A mathematical representation of patterns in data used for prediction or classification.

8. Training Data â€“ Data used to teach a machine learning model.

9. Test Data â€“ Data used to evaluate how well a model performs on unseen examples.

10. Overfitting â€“ When a model learns training data too well, including noise, and performs poorly on new data.

11. Underfitting â€“ When a model is too simple to capture the patterns in the data.

12. Accuracy â€“ The percentage of correct predictions made by a model.

13. Precision â€“ The proportion of true positive predictions among all positive predictions.

```python
#1. rename() method. Used to rename columns or index labels.

import pandas as pd

data = {
    "Blood_Pressure": [120, 140, 110],
    "Glucose_Level": [90, 110, 95]
}
df = pd.DataFrame(data)

# Rename columns
df.rename(columns={"Blood_Pressure": "BP", "Glucose_Level": "Glucose"}, inplace=True)

# Rename index
df.rename(index={0: "Patient1", 1: "Patient2"}, inplace=True)

print(df)


```

```python
# 2. loc[] vs iloc[]

# loc[] â†’ label-based indexing (row/column names).

# iloc[] â†’ integer-based indexing (row/column positions).

df = pd.DataFrame({"A": [10,20,30], "B": [40,50,60]}, index=["x","y","z"])

# loc by label
print(df.loc["x", "A"])   # 10

# iloc by position
print(df.iloc[0, 0])      # 10

# loc for multiple rows
print(df.loc[["x","z"], "B"])

# iloc for multiple rows
print(df.iloc[[0,2], 1])
```

```python
# 3. replace() method

# Used to replace values.

df = pd.DataFrame({"Status": ["fail", "pass", "fail", "pass"]})

# Replace single value
df["Status"] = df["Status"].replace("fail", "retry")

# Replace multiple values
df["Status"] = df["Status"].replace({"retry": "pending", "pass": "approved"})
```

```python
# 4. fillna() method

# Used to fill missing values (NaN).

df = pd.DataFrame({"A":[1, None, 3], "B":[4, None, 6]})

# Fill with constant
df.fillna(0, inplace=True)

# Fill column-wise
df["A"].fillna(df["A"].mean(), inplace=True)  # replace NaN with mean
```

```python
# 5. Conditional filtering

# Filter rows based on conditions.

df = pd.DataFrame({"Name": ["Anna","Ben","Chris"],
                   "Age": [22, 35, 19],
                   "Score": [85, 40, 90]})

# Single condition
print(df[df["Age"] > 20])

# Multiple conditions (AND)
print(df[(df["Age"] > 20) & (df["Score"] > 50)])

# OR condition
print(df[(df["Age"] < 20) | (df["Score"] < 50)])
```
